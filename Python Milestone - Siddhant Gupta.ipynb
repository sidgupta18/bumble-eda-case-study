### Outlier Handling Strategy

---

For columns with significant outliers like age and income, simply deleting rows can lead to loss of valuable data, especially when outliers make up a considerable portion of the dataset. Instead, a more balanced approach will be used to minimize distortion while preserving insights.

The plan is to:

- **Avoid deleting data** unless values are clearly invalid (e.g., age < 18 or negative income).
- **Use trimmed statistics** â€” calculate the mean and median using only the middle 80% of the data (i.e., removing the top and bottom 10%). This helps reduce the influence of extreme values when summarizing distributions.
- **Retain outliers** in some cases (e.g., high income or tall height) if they appear valid and meaningful.



```python
# IQR Method for detecting outliers in age
Q1_age = bumble['age'].quantile(0.25)
Q3_age = bumble['age'].quantile(0.75)
IQR_age = Q3_age - Q1_age

lower_age = Q1_age - 1.5 * IQR_age
upper_age = Q3_age + 1.5 * IQR_age

age_outliers = bumble[(bumble['age'] < lower_age) | (bumble['age'] > upper_age)]
print(f"Number of outliers in age: {len(age_outliers)}")
```

```python
# IQR Method for detecting outliers in height
Q1_height = bumble['height'].quantile(0.25)
Q3_height = bumble['height'].quantile(0.75)
IQR_height = Q3_height - Q1_height

lower_height = Q1_height - 1.5 * IQR_height
upper_height = Q3_height + 1.5 * IQR_height

height_outliers = bumble[(bumble['height'] < lower_height) | (bumble['height'] > upper_height)]
print(f"Number of outliers in height: {len(height_outliers)}")
```

### Part 2.1: Binning and Grouping

---

Binning continuous variables into categories helps simplify analysis and uncover trends across defined groups. For this step, we will group `age` and `income` into meaningful segments to better understand user demographics and economic distribution.

The `age` column will be binned into four groups:
- 18â€“25
- 26â€“35
- 36â€“45
- 46+

The `income` column will be grouped using quantile-based thresholds into:
- Low Income
- Medium Income
- High Income

---


```python
#Assigning age groups

bins = [18, 25, 35, 45, np.inf]
labels = ['18-25', '26-35', '36-45', '46+']

bumble['age_group'] = pd.cut(bumble['age'], bins=bins, labels=labels, right=True)
```

```python
# Income buckets
q1 = bumble['income'].quantile(0.25)
q3 = bumble['income'].quantile(0.75)

conditions = [
    bumble['income'] <= q1,
    (bumble['income'] > q1) & (bumble['income'] < q3),
    bumble['income'] >= q3
]

labels = ['Low', 'Medium', 'High']

bumble['income_group'] = np.select(conditions, labels)
```

### Part 2.2: Derived Features

---

Derived features help enhance the dataset by extracting new information from existing columns. These features often uncover hidden patterns or add useful dimensions to the analysis.

In this step, we'll create a new column called `profile_completeness`, which reflects the percentage of non-missing values for each user profile. This will help us understand how thoroughly users fill out their profiles and whether completeness varies across demographics. The idea here is to check if whether or not profile completedness will have any statistically significant impact on user engagment. 

---


```python
#custom function to check profile completness

def profile_complete_check(df):
    completeness = (1 - df.isnull().sum(axis=1) / df.shape[1]) * 100
    return completeness.round(2).sort_values(ascending=False)
```

```python
bumble['profile_completeness'] = profile_complete_check(bumble).sort_index()
```

### Recommendation

### Part 4.4: Signs and Personality

---

Usersâ€™ self-reported zodiac signs (`sign`) can offer insights into personality preferences or trends. While not scientifically grounded, analyzing this data helps explore fun and engaging patterns.

Questions:
- Create a pie chart showing the distribution of zodiac signs (`sign`) across the platform. Which signs are most and least represented? Is this the right chart? If not, replace with right chart.
- How does sign vary across gender and status? Are there noticeable patterns or imbalances?

---


```python
#First we need to clean the data

# list of valid signs
valid_signs = [
    "aries", "taurus", "gemini", "cancer", "leo", "virgo",
    "libra", "scorpio", "sagittarius", "capricorn", "aquarius", "pisces"
]


#custom cleaning function 

def clean_sign(sign):
    if pd.isnull(sign):
        return np.nan
    for zodiac in valid_signs:
        if zodiac in sign.lower():
            return zodiac.capitalize()
    return "Other"
```

```python
bumble['sign_cleaned'] = bumble['sign'].apply(clean_sign)
```

```python
sign_dist_clean = prct(bumble['sign_cleaned'])
sign_df_clean = sign_dist_clean.reset_index()
sign_df_clean.columns = ['sign', 'percentage']

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(data=sign_df_clean, x='sign', y='percentage')
plt.title('Cleaned Zodiac Sign Distribution')
plt.ylabel('Percentage of Users')
plt.xlabel('Zodiac Sign')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Sign vs Gender
sign_gender_cleaned = pd.crosstab(
    bumble['sign_cleaned'], bumble['gender'],
    normalize='columns'
).round(4) * 100

# Sign vs Status
sign_status_cleaned = pd.crosstab(
    bumble['sign_cleaned'], bumble['status'],
    normalize='columns'
).round(4) * 100

# Display tables
display(sign_gender_cleaned)
display(sign_status_cleaned)
```

### Key Findings:
- The original sign column had inconsistent and quirky values like "aries and it matters a lot", "taurus but it doesn't matter", etc. We cleaned this data by extracting the valid zodiac keyword and discarding suffixes.
- Post-cleaning, the distribution of zodiac signs is nearly even, with Leo, Gemini, and Libra slightly more common among users.
- Distribution by gender shows a fairly balanced representation across signs, with Leo males being the most common.

- When analyzing signs by relationship status, no major skew is observed â€” though a few like Leo had more â€œunknownâ€ responses, which could indicate less seriousness or incomplete profiles.

### Recommendations:
Since sign data is self-reported and not impactful for engagement directly, it can be used in light-hearted features like horoscope match filters or personality nudges â€” but should not influence matchmaking logic.

Encourage users to pick valid sign values by limiting input options to a dropdown list, which prevents noisy or inconsistent data.
